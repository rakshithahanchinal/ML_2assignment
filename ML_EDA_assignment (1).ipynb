```python
  #11 Identify and remove outliers in the km_driven column using the IQR method ?



import pandas as pd

# Load dataset (Replace 'your_dataset.csv' with actual filename)
df = pd.read_csv('BIKE DETAILS.csv')

# Check available columns
print(df.columns)

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

# Compute Interquartile Range (IQR)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"IQR: {IQR}, Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Identify outliers
outliers = df[(df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound)]
print(f"Number of outliers: {outliers.shape[0]}")

# Remove outliers
df_cleaned = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

print(f"Dataset after outlier removal: {df_cleaned.shape[0]} rows remaining")
```

```python
  #13   What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)?


import pandas as pd

# Load dataset (Replace 'your_dataset.csv' with actual filename)
df = pd.read_csv('BIKE DETAILS.csv')

# Check available columns
print(df.columns)

# Define current year dynamically
from datetime import datetime
current_year = datetime.now().year

# Calculate bike age
df['bike_age'] = current_year - df['year']

# Calculate depreciation (selling_price / bike_age)
df['depreciation'] = df['selling_price'] / df['bike_age']

# Calculate average depreciation
average_depreciation = df['depreciation'].mean()

print(f"Average depreciation in selling price based on bike's age: {average_depreciation:.2f} per year")
```

```python
   #14  Which bike names are priced significantly above the average price for their manufacturing year ?

import pandas as pd

# Load dataset (Replace 'your_dataset.csv' with actual filename)
df = pd.read_csv('BIKE DETAILS.csv')

# Check available columns
print(df.columns)

# Calculate the average selling price for each manufacturing year
avg_price_by_year = df.groupby('year')['selling_price'].mean()

# Merge average prices back to the original dataframe
df['avg_price_for_year'] = df['year'].map(avg_price_by_year)

# Define "significantly above average" as being 30% higher than the yearly average
threshold = 1.3  # 30% above average

# Filter bikes priced above the threshold
expensive_bikes = df[df['selling_price'] > threshold * df['avg_price_for_year']]

# Select relevant details
expensive_bikes_list = expensive_bikes[['name', 'year', 'selling_price', 'avg_price_for_year']]

print("Bikes priced significantly above the average price for their manufacturing year:")
print(expensive_bikes_list)
```

```python
    #12    Is there any seasonal trend in car sales based on the date of sale ?





import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (replace 'your_dataset.csv' with actual file path)
df = pd.read_csv("Car Sale (1).csv")

# Convert 'Date_of_Sale' to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Extract year and month for analysis
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month

# Aggregate sales by month to observe seasonal trends
monthly_sales = df.groupby("Month")["Price ($)"].sum().reset_index()

# Line plot: Monthly car sales trend
plt.figure(figsize=(10, 6))
sns.lineplot(x="Month", y="Price ($)", data=monthly_sales, marker="o", color="b")
plt.title("Seasonal Trend in Car Sales (By Month)")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(True)
plt.show()

# Boxplot: Sales distribution across months
plt.figure(figsize=(10, 6))
sns.boxplot(x="Month", y="Price ($)", data=df)
plt.title("Monthly Variation in Car Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(True)
plt.show()
```

```python
# 6  What are the most popular product keywords ?


import pandas as pd
from collections import Counter
import re

# Load dataset
df = pd.read_csv("amazon.csv")  # Replace with your actual dataset file path

# Function to extract keywords from product names
def extract_keywords(product_name):
    words = []
    for name in product_name.dropna():  # Remove missing values
        words.extend(re.findall(r'\b\w+\b', str(name).lower()))  # Tokenize and clean text
    return words

# Extract keywords
keywords = extract_keywords(df["product_name"])

# Count keyword frequency
keyword_counts = Counter(keywords)

# Convert to DataFrame for better readability
keyword_df = pd.DataFrame(keyword_counts.items(), columns=["Keyword", "Count"])

# Sort by most popular keywords
keyword_df = keyword_df.sort_values(by="Count", ascending=False)

# Display top keywords
print(keyword_df.head(10))  # Top 10 most frequent keywords
```

```python
#10    Identify any potential areas for improvement or optimization based on the data analysis ?




import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("amazon.csv")  # Replace with your actual dataset file path

# 1. Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values[missing_values > 0])

# 2. Detect outliers using IQR method
def detect_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[column] < Q1 - 1.5 * IQR) | (df[column] > Q3 + 1.5 * IQR)]
    return outliers

numeric_columns = df.select_dtypes(include=["number"]).columns
for col in numeric_columns:
    outliers = detect_outliers(col)
    print(f"Outliers detected in {col}: {len(outliers)}")

# 3. Analyze correlation between important variables
correlation_matrix = df.corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

# 4. Identify underperforming categories based on average ratings or sales
if "Category" in df.columns and "Rating" in df.columns:
    avg_rating_by_category = df.groupby("Category")["Rating"].mean().reset_index()
    lowest_rated_categories = avg_rating_by_category.sort_values(by="Rating", ascending=True).head(5)
    print("\nLowest Rated Categories:\n", lowest_rated_categories)

if "Category" in df.columns and "Quantity Sold" in df.columns:
    sales_by_category = df.groupby("Category")["Quantity Sold"].sum().reset_index()
    lowest_sales_categories = sales_by_category.sort_values(by="Quantity Sold", ascending=True).head(5)
    print("\nLowest Sales Categories:\n", lowest_sales_categories)

# 5. Identify potential pricing optimizations
if "Original Price" in df.columns and "Discounted Price" in df.columns:
    df["Discount Percentage"] = ((df["Original Price"] - df["Discounted Price"]) / df["Original Price"]) * 100
    avg_discount_by_category = df.groupby("Category")["Discount Percentage"].mean().reset_index()
    print("\nAverage Discount by Category:\n", avg_discount_by_category)
```

```python
#9   Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
#   violin plot


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("spotify.csv")  # Replace with your actual dataset file path

# Ensure relevant columns exist
if "Artist" in df.columns and "Duration (ms)" in df.columns:
    # Select top artists for better readability
    top_artists = df["Artist"].value_counts().index[:10]  # Select top 10 artists by track count
    df_filtered = df[df["Artist"].isin(top_artists)]

    # Set figure size
    plt.figure(figsize=(12, 6))

    # Create a box plot
    sns.boxplot(data=df_filtered, x="Artist", y="Duration (ms)")
    plt.xticks(rotation=45)
    plt.title("Track Duration Variation Across Top 10 Artists (Box Plot)")
    plt.xlabel("Artist")
    plt.ylabel("Track Duration (ms)")
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    plt.show()

    # Create a violin plot
    plt.figure(figsize=(12, 6))
    sns.violinplot(data=df_filtered, x="Artist", y="Duration (ms)", inner="quartile")
    plt.xticks(rotation=45)
    plt.title("Track Duration Variation Across Top 10 Artists (Violin Plot)")
    plt.xlabel("Artist")
    plt.ylabel("Track Duration (ms)")
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    plt.show()

else:
    print("Dataset does not contain required columns: 'Artist' and 'Duration'.")
```
